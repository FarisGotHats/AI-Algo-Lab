## 2. Discrete Distribution: Bernoulli
#### **Characteristics**: Binary outcomes (0/1, Success/Failure)

### Bernoulli Distribution: Foundation of Binary Outcomes

#### Definition
The **Bernoulli distribution** is a discrete probability distribution modeling experiments with **exactly two mutually exclusive outcomes**:
- Success (1) with probability `p`
- Failure (0) with probability `1-p`

**Probability Mass Function (PMF):**  
`P(X = x) = p^x(1-p)^(1-x)`  
where:  
- `x ‚àà {0, 1}`  
- `0 ‚â§ p ‚â§ 1` (probability of success)

#### Key Characteristics
- **Discrete**: Only defined for integer values 0 and 1  
- **Memoryless**: Each trial is independent  
- **Fundamental Building Block**: Forms basis for Binomial distribution (multiple Bernoulli trials)

#### AI/ML Applications
- üõ°Ô∏è **Spam Detection**: Classify emails as spam (1) or not spam (0)  
- üéØ **Click Prediction**: Estimate probability of user clicking an ad  
- üß™ **Medical Testing**: Predict positive/negative diagnostic results  
- ü§ñ **Activation Functions**: Basis for binary classifiers in neural networks

**Why It Matters in AI:**  
Provides mathematical foundation for:  
- Logistic regression classifiers  
- Binary cross-entropy loss functions  
- Bernoulli Naive Bayes algorithms  
- Threshold-based decision making

#

```python
# Bernoulli Distribution Parameters
p = 0.3  # Probability of success (e.g., email is spam)
size = 1000  # Number of trials

# Generate Bernoulli data
np.random.seed(42)
data_bernoulli = bernoulli.rvs(p=p, size=size)

# Calculate PMF (Probability Mass Function)
x_values = [0, 1]
pmf_values = bernoulli.pmf(x_values, p)
```

```python
# Visualization
plt.bar(x_values, pmf_values, color=['skyblue', 'salmon'], alpha=0.7)
plt.title(f'Bernoulli Distribution (p={p})')
plt.xticks(x_values, ['Not Spam (0)', 'Spam (1)'])
plt.ylabel('Probability') # Probability of the Spam/Non Spam
plt.xlabel('Outcomes')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data_bernoulli, bins=[-0.5, 0.5, 1.5], edgecolor='black', density=True, color='lightgreen')
plt.xticks([0, 1], ['Not Spam (0)', 'Spam (1)'])
plt.xlabel('Outcomes')
plt.ylabel('Frequency')
plt.title(f'Bernoulli Distribution Simulation (p={p}, size={size}) - Histogram')
plt.show()
```

```python
# --- Alternative Visualization 2:  Scatter Plot with Jitter ---
plt.figure(figsize=(8, 6))
y_jitter = np.random.normal(0, 0.05, size=size)  # Add some vertical jitter
plt.scatter(data_bernoulli, y_jitter, alpha=0.3, color='purple')
plt.yticks([])  # Hide y-axis ticks
plt.xticks([0, 1], ['Not Spam (0)', 'Spam (1)'])
plt.xlabel('Outcomes')
plt.title(f'Bernoulli Distribution Simulation (p={p}, size={size}) - Scatter Plot with Jitter')
plt.show()
```

### AI Use Case:
#### Simplified example using logistic regression

```python
# Features: [urgent, money, click]
X = np.array([
    [0, 0, 1],  # No 'urgent', No 'money', Has 'click'
    [1, 1, 0],  # Has 'urgent', Has 'money', No 'click'
    [0, 1, 1],
    [1, 0, 0],
    [0, 0, 0]
])

# Corresponding labels (0 = not spam, 1 = spam)
y = np.array([0, 1, 1, 1, 0])
```

```python
# Train classifier
model = LogisticRegression()
model.fit(X, y)
```

```python
# Predict new email
new_email = np.array([[1, 1, 1]])  # Contains all spam keywords
prob = model.predict_proba(new_email)[0][1]
print(f"Probability email is spam: {prob:.2%}")
```
